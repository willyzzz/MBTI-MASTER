```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
import time
import datetime
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
```

```python
#判断元素是否存在
def isElementExist_fast(xpath):
    try:
        driver.find_element_by_xpath(xpath)
        return True
    except:
        return False

def isElementExist(xpath):
    try:
        wait.until(lambda x: x.find_element_by_xpath(xpath))
        return True
    except:
        return False
```

```python
#打开chrome浏览器
chrome_options = Options()
chrome_driver = r"E:\develop\pythonproject\chromedriver.exe"
driver = webdriver.Chrome(chrome_driver, chrome_options= chrome_options)
driver.maximize_window()

wait = WebDriverWait(driver,5)
wait_long = WebDriverWait(driver,10)

#名人
driver.get('https://www.personality-database.com/profile?pid=1&sort=top')
#虚拟角色
#driver.get('https://www.personality-database.com/profile?pid=2&sort=alphabet')

#关闭广告
#advertise = wait_long.until(lambda x: x.find_element_by_xpath('//*[@id="ezmobfooter"]/span'))
#advertise.click()
```

```python
#chrome打开后，手动在页面下方将一页显示数量改成500，并将广告和cookie关掉
```

```python
original_window = driver.current_window_handle
datalist = []
#在此设置一次抓取的页数
totalpage = 1

for page in range(totalpage):
    for i in range(500):
        #获取评论数看是否为0
        comm = wait.until(lambda x: x.find_element_by_xpath("//*[@id='root']/div/section/main/div[1]/div[2]/div/div[5]/a["+str(i+1)+"]/div/div[3]/div[1]/label")).text
        commnum = comm.split('/')[1]
        if commnum != '0':
            #获取姓名
            name = wait.until(lambda x: x.find_element_by_xpath("//*[@id='root']/div/section/main/div[1]/div[2]/div/div[5]/a["+str(i+1)+"]/div/div[2]/h2")).text
            print(name)
            #滚动到当前位置并在新标签页打开子网页
            item = wait.until(lambda x: x.find_element_by_xpath("//*[@id='root']/div/section/main/div[1]/div[2]/div/div[5]/a["+str(i+1)+"]/div"))
            driver.execute_script("window.scrollTo(0,0)")
            ActionChains(driver).move_to_element(item).perform()
            ActionChains(driver).key_down(Keys.CONTROL).click(item).key_up(Keys.CONTROL).perform()
            driver.switch_to.window(driver.window_handles[-1])
            try:
                #滑动到评论区
                item = wait_long.until(lambda x: x.find_element_by_xpath("//*[@id='root']/div/section/main/div[1]/div[2]/div[1]/div[1]/div[1]/div[5]/div[1]/span[1]"))
                ActionChains(driver).move_to_element(item).perform()
                #关闭广告
                advertise = wait.until(lambda x: x.find_element_by_xpath('//*[@id="ezmobfooter"]/span'))
                advertise.click()
                #关闭cookie提醒
                if isElementExist_fast('//*[@id="rcc-confirm-button"]'):
                    cookie = wait.until(lambda x: x.find_element_by_xpath('//*[@id="rcc-confirm-button"]'))
                    cookie.click()
                #点击New标签
                item.click()

                #获取评论
                j = 1
                while 1:
                    if isElementExist("//*[@id='root']/div/section/main/div[1]/div[2]/div[1]/div[1]/div[1]/div[5]/div[2]/div["+str(j)+"]/div/div[2]/div[1]/div") == False:
                        #print('a '+str(j))
                        break
                    review = wait.until(lambda x: x.find_element_by_xpath("//*[@id='root']/div/section/main/div[1]/div[2]/div[1]/div[1]/div[1]/div[5]/div[2]/div["+str(j)+"]/div/div[2]/div[1]/div")).text
                    if isElementExist_fast("//*[@id='root']/div/section/main/div[1]/div[2]/div[1]/div[1]/div[1]/div[5]/div[2]/div["+str(j)+"]/div/span") == False:
                        label = ''
                    else:
                        label = wait.until(lambda x: x.find_element_by_xpath("//*[@id='root']/div/section/main/div[1]/div[2]/div[1]/div[1]/div[1]/div[5]/div[2]/div["+str(j)+"]/div/span")).text
                    item = [name,label,review]
                    datalist.append(item)

                    j = j + 1
                    if j in [3,4,9,10,11] and isElementExist_fast("//*[@id='root']/div/section/main/div[1]/div[2]/div[1]/div[1]/div[1]/div[5]/div[2]/div["+str(j)+"]/div/div[2]/div[1]/div") == False:
                        j = j + 1
                    if isElementExist_fast("//*[@id='root']/div/section/main/div[1]/div[2]/div[1]/div[1]/div[1]/div[5]/div[2]/div["+str(j)+"]/div/div[2]/div[1]/div") == False:
                        driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
                        time.sleep(2)
                        #print('b '+str(j))
                driver.close()
                driver.switch_to.window(original_window)
                time.sleep(1)
            except:
                driver.close()
                driver.switch_to.window(original_window)
                time.sleep(1)
    #点击下一页
    nextpage = wait.until(lambda x: x.find_element_by_xpath("//*[@id='root']/div/section/main/div[1]/div[2]/div/div[5]/ul/li[9]/button"))
    ActionChains(driver).move_to_element(nextpage).perform()
    nextpage.click()
    time.sleep(2)
```

```python
print(len(datalist))
import pandas as pd
df = pd.DataFrame(datalist)
df.to_csv('famous_top_500.csv',index = False,header = False)
```
