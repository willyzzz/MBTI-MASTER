```python
import pandas as pd
df = pd.read_csv('preprocess_data.csv')
```

```python
df.dropna(axis=0,inplace=True)
```

```python
w = []
for i in df['Type']:
    s = i.split(' - ')
    w.append(s)
```

```python
df['Type'] = w
```

```python
pers = ['INTJ','INTP','INFJ','INFP','ISTJ','ISTP','ISFJ','ISFP','ENTJ','ENTP','ENFJ','ENFP','ESTJ','ESTP','ESFJ','ESFP']
```

```python
q = []
for i in df['Type']:
    for val in i:
        if val in pers:
            q.append(val)
        else:
            q.append(0)
        break

df.index = range(len(df))
df['Type'] = q
```

```python
q = []
for i in df['Text']:
    i = eval(i)
    q.append(i)
df['Text'] = q
df.index = range(len(df))
```

```python
df.dropna(axis=0,inplace=True)
```

```python
#length restriction
text = df['Text']

for i in range(len(text)):
    if len((text[i])) < 20 :
        df = df.drop(i)
df.index = range(len(df))
```

```python
data = df.drop(index = df['Type'][df['Type'] == 0].index)
```

```python
data.index = range(len(data))
data.to_csv('preprocess_data_final.csv')
```

```python
import pandas as pd
df = pd.read_csv('preprocess_data_final.csv')
```

```python
q = []
for i in df['Text']:
    i = eval(i)
    q.append(i)
df['Text'] = q
df.index = range(len(df))
```

```python
def stopwords_remove(sentence):
    #stoplist = stopwords.words('english')
     
    with open('C:/Users/willy/Desktop/stopwords.txt') as file:
        stoplist = [stopword.replace('\n', '').lower() for stopword in file.readlines()]
    
    new_sentence = [word for word in sentence if word not in stoplist]
    return new_sentence
```

```python
new_sentence_stopwords = df['Text'].apply(lambda text: stopwords_remove(text))
```

```python
new_sentence_stopwords = [word for word in new_sentence_stopwords if word != '”']
```

```python
df['Text'] = new_sentence_stopwords
```

```python
def remove_special(text):
    text = [i for i in text if '\x88'and '\x8c' and '\x83' and '”' and '\x9b' and 'é' and '¥æª' and 'é' and '”' and '¥æª' not in i]
    return text
```

```python
text = df['Text'].apply(lambda x:remove_special(x))
```

```python
df = df.loc[df.Text.apply(lambda x: not isinstance(x, (float, int)))]
```

```python
df['Text'] = text
```

```python
df.index = range(len(df))
```
